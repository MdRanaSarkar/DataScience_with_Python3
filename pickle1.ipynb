```python
#For generating Pickle file
import pickle
import numpy as np
```

```python
data={
    'doc1':np.random.random(10),
    'next1':np.random.random(20)
}
```

```python
data
```

```python
with open('data.pkl','wb') as pickle_file:
    pickle.dump(data,pickle_file)
```

```python
with open('data.pkl','rb') as pickle_load:
    load=pickle.load(pickle_load)
```

```python
load
```
