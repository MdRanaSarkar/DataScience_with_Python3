```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```
df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/carprices.csv')
```

```
df.head()
```

```
dummies=pd.get_dummies(df['Car Model'])
```

```
#Now Read the dummies data
dummies
```

```
#Concatenating the dummy variable
marged=pd.concat([df,dummies],axis='columns')
```

```
marged
```

```
final_dta=marged.drop(["Car Model"],axis='columns')
```

```
final_dta
```

```
#Put the predict data to y variable
y=df['Sell Price']
```

```
y
```

```
#Now droping the predicted value price
x=final_dta.drop(['Sell Price'],axis='columns')
```

```
x
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```
model.fit(x,y)
```

```
model.score(x,y)
```

```
model.predict([[52000,6,1,0,0,0]])
```
