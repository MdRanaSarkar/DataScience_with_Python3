```python
import keras
from keras.models import Sequential
from keras.layers.core import Dense,Dropout,Activation,Flatten
from keras.layers.convolutional import Convolution2D,MaxPooling2D
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('mnist_train.csv')
print(data.head())
```

```python
data
```

```python
r=data.iloc[3,1:].values.reshape(28,28).astype('uint8')
print(r)
```

```python
df_x=data.iloc[:,1:].values.reshape(len(data),28,28,1)
y=data.iloc[:,0].values
print(y)
```

```python
#converting lebel to categorical class
df_y=keras.utils.to_categorical(y,num_classes=10)
```

```python
df_x=np.array(df_x)
df_y=np.array(df_y)
print(df_y)
```

```python
print(df_y.shape)
```

```python
x_train,x_test,y_train,y_test=  train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
model=Sequential()
model.add(Convolution2D(32,3,data_format='channels_last',activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(Dense(10))
model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics=['accuracy'])
```

```python
model.summary()
```

```python
#for fitting it just 100 images for testing 
model.fit(x_train[1:100],y_train[1:100],validation_data=(x_test[1:20],y_test[1:20]))
```

```python
model.predict(x_test[1:20])
```
